## 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

相關性是需要任兩欄位之間作比較，所以現在若有 10 個欄位，那應該會有 10*10 個比較結果。
數學上很容易用矩陣表示，稱為相關性矩陣，這個很適合用 heatmap 的圖表來做呈現。

```python
##先做一個10*10的隨機矩陣，之後再去畫圖
matrix=np.random.random((10,10))*2-1 ##np.random() 是0~1之間，題目現在要-1~1之間   ,uniform() 是隨機生成實數在 [x, y] 範圍內
# matrix=np.random.uniform(-1,1,(10,10))
matrix=pd.DataFrame(matrix)
matrix
```

```python
## 畫相關性矩陣
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, annot = True)
"""
cmap=colormap 名稱或顏色對象，通常分成兩種，cubehelix map (數據為連續資料時) 或 RdBu_r (數據為離散資料時)
center:設立圖例中的均值，透過設置center值，可以調整的圖像顏色的整體深淺，亦可以增加vmax、vmin
annot=annotate的縮寫，annot正常默認為False，當annot為True時，在heatmap中每個方格中就可以寫入數據
"""

plt.show()
```

## 題2.3.條件隨機數值列, 並仿造進階 Heatmap 範例作圖

## 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.random((nrow,ncol))*2-1

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

###畫圖設定
# 上半部為 scatter
grid.map_upper(plt.scatter , alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.kdeplot )

# 下半部放 density plot
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

## 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(0,1,(nrow,ncol)) ##常態分佈的矩陣

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

###畫圖設定
# 上半部為 scatter
grid.map_upper(plt.scatter , alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.kdeplot )

# 下半部放 density plot
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
